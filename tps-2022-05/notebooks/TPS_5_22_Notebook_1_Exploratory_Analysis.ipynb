# TPS 05-22 - Exploratory Analysis

In this notebook, I want to test out Colab with kaggle competition data. The data has already been compressed significantly by downcasting datatypes and saving in the .feather format. This new data can be found in [my GitHub repository](https://github.com/rsizem2/tabular-playground-series). 

The EDA itself largely follows [this great notebook](https://www.kaggle.com/code/ambrosm/tpsmay22-eda-which-makes-sense).

```python
# Imports
import pandas as pd
import numpy as np
import pyarrow
import time
import re
import math
import gc

import plotly.express as px
import matplotlib.pyplot as plt
from IPython.display import Image
import seaborn as sns; sns.set_theme()
```

# Competition Data

```python
train = pd.read_feather('https://raw.githubusercontent.com/rsizem2/tabular-playground-series/main/tps-2022-05/data/train.feather')
```

```python
train.head()
```

```python
columns = [col for col in train.columns if col not in ['id','target']]

train[columns].info()
```

## Target Distribution

```python
fig = px.pie(
    train[['target','id']], 
    values = 'target',
    title='Target Values', 
)

fig.show()
```

## Float Distribution

```python
float_features = [f for f in train.columns if train[f].dtype == 'float32']

# Training histograms
fig, axs = plt.subplots(4, 4, figsize=(16, 16))
for f, ax in zip(float_features, axs.ravel()):
    ax.hist(train[f], density=True, bins=100)
    ax.set_title(f'Train {f}, std={train[f].std():.1f}')
plt.suptitle('Histograms of the float features')
plt.show()
```
